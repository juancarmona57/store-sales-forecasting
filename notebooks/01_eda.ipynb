# Store Sales - Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
DATA_DIR = Path('../data/raw')

train = pd.read_csv(DATA_DIR / 'train.csv', parse_dates=['date'])
stores = pd.read_csv(DATA_DIR / 'stores.csv')
oil = pd.read_csv(DATA_DIR / 'oil.csv', parse_dates=['date'])
holidays = pd.read_csv(DATA_DIR / 'holidays_events.csv', parse_dates=['date'])
transactions = pd.read_csv(DATA_DIR / 'transactions.csv', parse_dates=['date'])

print('train:', train.shape)
print('stores:', stores.shape)
print('oil:', oil.shape)
print('holidays:', holidays.shape)
print('transactions:', transactions.shape)
```

```python
print('=== train.info() ===')
train.info()
print('\n=== train.describe() ===')
display(train.describe())
```

```python
print('=== Missing values per column ===')
for df_name, df in [('train', train), ('stores', stores), ('oil', oil), ('holidays', holidays), ('transactions', transactions)]:
    nulls = df.isnull().sum()
    if nulls.any():
        print(f'\n{df_name}:')
        print(nulls[nulls > 0])
    else:
        print(f'\n{df_name}: no missing values')
```

```python
plt.figure(figsize=(10, 4))
plt.hist(train['sales'], bins=50, edgecolor='black', color='steelblue')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
zero_pct = (train['sales'] == 0).mean()
print(f'Zero sales: {zero_pct:.1%}')
print(f'Total rows: {len(train):,}')
print(f'Rows with zero sales: {(train["sales"] == 0).sum():,}')
```

```python
top10_families = train.groupby('family')['sales'].mean().nlargest(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top10_families.values, y=top10_families.index, palette='viridis')
plt.title('Top 10 Product Families by Mean Sales')
plt.xlabel('Mean Sales')
plt.ylabel('Family')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = train.groupby(train['date'].dt.to_period('M'))['sales'].mean()
monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(14, 4))
monthly_sales.plot(title='Monthly Sales Trend', color='darkorange')
plt.xlabel('Month')
plt.ylabel('Mean Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 4))
plt.plot(oil['date'], oil['dcoilwtico'], color='green', linewidth=0.8)
plt.title('Oil Price Over Time (WTI)')
plt.xlabel('Date')
plt.ylabel('Oil Price (USD/barrel)')
plt.tight_layout()
plt.show()
```
